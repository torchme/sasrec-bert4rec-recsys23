```python
import pandas as pd
```

```python
def get_sum_seqs(seqs):
    s = 0
    for seq in seqs.values():
        s += len(seq)
    return s


def get_items(seqs):
    res = dict()
    for seq in seqs.values():
        for item in seq:
            res[item] = res.get(item, 0) + 1
    return res
```

```python
train_seqs = pd.read_pickle('cold_item_train_sequences.pkl')
print(get_sum_seqs(train_seqs))
train_items2count = get_items(train_seqs)
len(train_seqs)
```

```python
val_seqs = pd.read_pickle('valid_sequences.pkl')
val_items2count = get_items(val_seqs)
print(get_sum_seqs(val_seqs))
len(val_seqs)
```

```python
test_seqs = pd.read_pickle('test_sequences.pkl')
test_items2count = get_items(val_seqs)
print(get_sum_seqs(test_seqs))
len(test_seqs)
```

```python
with_cold_inside = []
for user, seq in test_seqs.items():
    for item in seq:
        if item not in train_items2count:
            with_cold_inside.append(user)
            break
len(with_cold_inside)
```
